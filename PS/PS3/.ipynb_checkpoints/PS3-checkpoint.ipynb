```python
# import libraries
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
# Show plots in the notebook
%matplotlib inline
```

```python
#import data
data = xr.open_dataset("NOAA_NCDC_ERSST_v3b_SST.nc", engine="netcdf4")
# check data
data
```

```python
# slice the area needed (5N-5S, 170W-120W)
data = data.sel(lat=slice(-5,5), lon=slice(190, 240))
# check data
data
```

```python
# check null values
data.isnull().sum()
```

```python
# calculate the mean of the data according to time
mean_time = data.mean(dim=['lat', 'lon'])
#check mean_time
mean_time
```

```python
#calcualate the mean of the data according to month
month_time = mean_time.groupby('time.month').mean()
month_time
```

```python
# calculate the anomalies
anomalies = mean_time.groupby('time.month') - month_time
# check anomalies
anomalies
```

```python
# plot the anomalies
plt.figure(figsize=(8,10))
anomalies.sst.plot()
```

```python
plt.figure(figsize=(5,5), dpi=1000)
plt.bar(anomalies['time'], anomalies['sst'])
```
