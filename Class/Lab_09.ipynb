### 1. Interactive visualization using hvplot

```python
# Import modules
import numpy as np
import pandas as pd
import xarray as xr
from matplotlib import pyplot as plt
%matplotlib inline
%matplotlib qt5
```

```python

# Open the netCDF4 file
ds = xr.open_dataset("sst.mnmean.nc", engine="netcdf4")

# Check the data
ds
```

```python
import hvplot.xarray
```

```python
# Show all data
ds.sst.hvplot()
```

```python
# Show 2-D map
ds.sst.isel(time=-1).hvplot(cmap="fire")
```

```python
# Show 1-D line
ds.sst.isel(time=-1, lon=100).hvplot()
```

```python
# Show 1-D line
ds.sst.sel(lon=114.55, lat=22.5, method='nearest').hvplot()
```

```python
# Show widgets
ds.sst.hvplot(groupby="time", clim=(ds.sst.min(), ds.sst.max()), cmap='turbo')
```

```python
# Add more features to the time widget
ds.sst.hvplot(
    groupby="time",
    clim=(ds.sst.min(), ds.sst.max()),
    cmap="turbo",
    widget_type="scrubber",
    widget_location="bottom",
)
```
